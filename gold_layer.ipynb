### Data Reading , Writing and Creating delta tables

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
silver = 'abfss://silver@nyctaxistorageac.dfs.core.windows.net'
gold = 'abfss://gold@nyctaxistorageac.dfs.core.windows.net'
```

**Storage Variables**

Trip Zone

```python
df_zone = spark.read.format('parquet')\
    .option('inferSchema' , True)\
        .option('header' , True)\
            .load(f'{silver}/trip_zone')
```

```python
df_zone.display()
```

```python
df_zone.write.format('delta')\
             .mode('append')\
             .option('path', f'{gold}/trip_zone')\
             .saveAsTable('gold.trip_zone')
     
```

```python
%sql
SELECT * FROM gold.trip_zone;
```

Trip Type

```python
df_type = spark.read.format('parquet')\
    .option('inferSchema' , True)\
        .option('header' , True)\
            .load(f'{silver}/trip_type')
```

```python
df_type.write.format('delta')\
             .mode('append')\
             .option('path', f'{gold}/trip_type')\
             .saveAsTable('gold.trip_type')
```

Trip Data

```python
df_trip = spark.read.format('parquet')\
    .option('inferSchema' , True)\
        .option('header' , True)\
            .load(f'{silver}/trip2023data')
```

```python
df_trip.write.format('delta')\
             .mode('append')\
             .option('path', f'{gold}/trip2023data')\
             .saveAsTable('gold.trip2023data')
```
